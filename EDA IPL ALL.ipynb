Highest Total run by Team of IPL

```python
 batting_summary.groupby("Team Batting")["Total Runs"].sum().sort_values(ascending=False).reset_index().head(5)
```

```python
total=batting_summary.groupby("Team Batting")["Total Runs"].sum().sort_values(ascending=False).reset_index().head(5)
sns.barplot(x=total["Team Batting"],y=total["Total Runs"],data=total,palette="spring")
for i, v in enumerate(total["Total Runs"]):
    plt.text(i,v,str(v),ha="center",va="bottom")
plt.xticks(rotation=45)
plt.title("Top 5 Total Run Score by Team")
plt.show()
```

Mumbai Indians is a highest run scorer team

Highest Score of singel innings in IPL

```python
batting_summary.groupby("Team Batting")["Total Runs"].max().sort_values(ascending=False).reset_index().head(5)
```

```python
highest_run=batting_summary.groupby("Team Batting")["Total Runs"].max().sort_values(ascending=False).reset_index().head(5)
sns.barplot(x=highest_run["Team Batting"],y=highest_run["Total Runs"],data=highest_run,palette="summer")
plt.title("Highest Score in a single innings IPL history ever")
plt.xticks(rotation=45)
for i,v in enumerate(highest_run["Total Runs"]):
    plt.text(i,v,str(v),ha="center",va="bottom")
plt.show()
```

Royal Challengers Bangalore is a highest scorer of the single innings

```python
#Data Distribution
sns.boxplot(bowling_summary[["Total Wickets Taken","Bowler Name","Overs","Maidens","Runs","Wickets","No Balls","Wides","Economies"]])
plt.xticks(rotation=45)
plt.ylabel("Frequency")
plt.title("Data Distribution of Numeric Variable")
plt.show()
```

```python
#Exploring Numerical Variable
total_wicket=bowling_summary.groupby("Bowler Name")["Wickets"].sum().sort_values(ascending=False).reset_index().head(10)
sns.barplot(x=total_wicket["Bowler Name"],y=total_wicket["Wickets"],data=total_wicket,hue="Bowler Name")
plt.xticks(rotation=45)
plt.title("Top 10 Bolwer by total taken wickets in IPL 2008-2023")
plt.show()
```

```python
last_year=bowling_summary[bowling_summary["IPL Edition"]==2023]
bowler=last_year.groupby("Bowler Name")["Wickets"].sum().sort_values(ascending=False).reset_index().head(10)
sns.barplot(x=bowler["Bowler Name"],y=bowler["Wickets"],data=bowler,hue="Bowler Name")
plt.xticks(rotation=45)
plt.title("Top Bowler by wicket taken 2023")
plt.show()
```

```python
over=bowling_summary.groupby("Bowler Name")["Overs"].sum().sort_values(ascending=False).reset_index().head(10)
sns.barplot(x=over["Bowler Name"],y=over["Overs"],data=over,hue="Bowler Name")
plt.xticks(rotation=45)
plt.title("Total over by Bowler 2008-2023")
plt.show()
```

```python
maiden=bowling_summary.groupby("Bowler Name")["Maidens"].sum().sort_values(ascending=False).reset_index().head(10)
sns.barplot(x=maiden["Bowler Name"],y=maiden["Maidens"],data=maiden,hue="Bowler Name")
plt.xticks(rotation=45)
plt.title("Highest Maidens over by Bowler 2008-2023")
plt.show()
```

```python
Economies=bowling_summary.groupby("Bowler Name")["Economies"].mean().sort_values(ascending=False).reset_index().head(10)
sns.barplot(x=Economies["Bowler Name"],y=Economies["Economies"],data=Economies,hue="Bowler Name")
plt.xticks(rotation=45)
plt.title("Highest Economies by Bowler 2008-2023")
plt.show()
```

```python
co_relation=bowling_summary[["IPL Edition","Total Wickets Taken","Overs","Maidens","Runs","Wickets","No Balls","Wides","Economies"]].corr()
sns.heatmap(co_relation,annot=True,cmap="coolwarm",fmt="0.2f")
plt.title("co-relation of bowling_summary")
plt.show()
```
