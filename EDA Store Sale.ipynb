```python
#Relationship of total sale and own store
```

```python
def ownstore_to_own_store(x):
    if x==1:
        return "Own_Stores"
    else:
        return "Others_Stores"
```

```python
df["Own_store"]=df["OwnStore"].apply(ownstore_to_own_store)
```

```python
store_sale=df.groupby("Own_store")["TotalSales"].sum().sort_values(ascending=False).reset_index()
store_sale 
```

```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
sns.boxplot(x="Own_store",y="TotalSales",data=df,hue="Own_store")
plt.title("Total Sale vs Own Store")

plt.subplot(2,2,2)
plt.pie(store_sale["TotalSales"],labels=store_sale["Own_store"],autopct="%0.f%%")
plt.title("Total Sale vs Own Store and Other Store")
plt.legend()
plt.show()
```

Observation:
* above the graph we analyze other store much more sale than own store.

```python
def online_presence(x):
    if x==1:
        return "Online"
    else:
        return "Offline"
```

```python
#Relation of online presence and total sales
df["online_presence"]=df["OnlinePresence"].apply(online_presence)
```

```python
online=df.groupby("online_presence")["TotalSales"].sum().sort_values(ascending=False).reset_index()
online
```

```python
sns.barplot(x=online["online_presence"],y=online["TotalSales"],data=df,hue=online["online_presence"])
for i,v in enumerate(online["TotalSales"]):
    plt.text(i,v,str(v),ha="center",va="bottom")
plt.title("Online Presence vs Total Sales")
plt.show()
```

Observation:
* above the graph we analyze most of sale is off-line.

Key Insights:

Store Distribution:

Apparel Zone emerges as the dominant store type, with equal representation in Delhi, Mumbai, and Kolkata.

The dataset reveals a balanced distribution of stores across these major cities, indicating a competitive retail landscape.
Operating Costs and Sales:

A significant portion of stores operates with operating costs ranging from 50 to 150, correlating with moderate sales figures.
Customer Engagement:

Apparel stores attract the maximum number of customers, suggesting a strong appeal and demand for apparel products among consumers.
Profitability Analysis:

Apparel stores emerge as the most profitable, indicating a lucrative market for fashion retail.
Comparative Sales:

Electronics and apparel stores outperform supermarkets in terms of sales volume, highlighting consumer preferences and market dynamics.
Top Performing Store:

Fashion Bazar emerges as the highest sales-generating store, with Delhi and Mumbai as its primary markets.
Sales Channels:

Offline sales dominate the retail landscape, indicating the enduring importance of brick-and-mortar stores despite the rise of e-commerce.
