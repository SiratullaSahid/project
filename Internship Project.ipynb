```python
#analysing others stores and own stores sales
```

```python
def ownstore(x):
    if x==1:
        return "Own_Store"
    else:
        return "Other_Store"
```

```python
df["Own_Store"]=df["OwnStore"].apply(ownstore)
```

```python
store=df["Own_Store"].value_counts().reset_index()
store
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.countplot(x=df["Own_Store"],palette="spring")
plt.title("Count of Stores")
for i,v in enumerate(store["count"]):
    plt.text(i,v,str(v),ha="center",va="bottom")

plt.subplot(1,2,2)
plt.pie(store["count"],labels=store["Own_Store"],autopct="%0.1f%%")
plt.title("Percentage of Store Type")
plt.show()
```

Observation:
    1. 18th Others stores and 14th own stores.
    2. 56% Store is Others stores and 43% own stores.

```python
cost=df.groupby("Own_Store")["TotalSales"].sum().reset_index()
cost
```

```python
df.groupby("Own_Store")[["ProfitPercust","TotalSales"]].sum().reset_index()
```

```python
df.groupby("Own_Store")[["Staff_Cnt","OperatingCost"]].sum().reset_index()
```

```python
acqcost=df.groupby("Own_Store")["AcqCostPercust"].sum().reset_index()
acqcost
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.pie(cost["TotalSales"],labels=cost["Own_Store"],autopct="%0.1f%%")
plt.title("Total Sale vs Stores Type")
plt.subplot(2,2,4)
sns.barplot(y=acqcost["AcqCostPercust"],hue=acqcost["Own_Store"])
plt.xlabel("AcqCostPercust vs Stores Type")
plt.subplot(2,2,3)
sns.lineplot(x=df["Staff_Cnt"],y=df["OperatingCost"],hue=df["Own_Store"])
plt.title("Operating Cost and Staff Cnt by Stores Type")
plt.subplot(2,2,2)
sns.lineplot(x=df["TotalSales"],y=df["ProfitPercust"],hue=df["Own_Store"])
plt.title("Profit and Total Sales by Stores Type")
plt.show()
```

Observation:
    1. 75% sale from others stores.
    2. Own stores have more profitable than others stores.
    3. if staff has been incrised then operating cost has been decresed.
    4. Others stores acq-cost is higher than own stores.

Observation:
    1. 75% sale from offline.
    2. Online Stores operating cost higher than offline Stores.
    3. Most of coustomers have online Stores.
    4. Off-line store has more staff than online store.
    5. Off-line store is more profitable than online store.
    6. Online stores Acq-cost is high than offline store.

```python
#Co-Relation Between total sale vs others variable
plt.figure(figsize=(17,15))
plt.subplot(3,3,1)
sns.lineplot(x="TotalSales",y="OperatingCost",hue="Staff_Cnt",data=df)
plt.title("Operating Cost vs Total Sale by StaffCnt")
plt.subplot(3,3,2)
sns.scatterplot(x=df["Total_Customers"],y=df["TotalSales"],color="r")
plt.title("Total Coustomer vs Total Sale")
plt.subplot(3,3,3)
sns.lineplot(x=df["TotalSales"],y=df["AcqCostPercust"])
plt.title("Total Sale vs Acq Cost")
plt.subplot(3,3,7)
sns.scatterplot(x=df["TotalSales"],y=df["BasketSize"],color="g")
plt.xticks(rotation=90)
plt.title("Total Sale vs Basket Size")
plt.subplot(3,3,5)
profit=df.groupby("StoreName")["ProfitPercust"].sum().reset_index()
plt.pie(profit["ProfitPercust"],labels=profit["StoreName"],autopct="%0.1f%%")
plt.title("Profit vs Store Name")
plt.subplot(3,3,6)
sns.barplot(y=df["TotalSales"],x=df["Tenure"],hue=df["Tenure"])
plt.title("Total Sale vs Tenure")
plt.subplot(3,3,4)
sns.lineplot(x=df["StoreSegment"],y=df["TotalSales"])
plt.title("Total Sale vs Store Segment")
plt.show()
```

Observation:
    1. if Total Sale is high then operating cost is low.
    2. staff have been incresed then sales incresed.
    3. Coustomers have been incresed then Sales have been incresed.
    4. Sale is high then Acq-cost will be low.
    5. Store segment is incresed then sale will be incresed.
    6. Apparel Zone is highest profitable store than others stores.
    7.Tenure is low then sale will be high.
    8. Basket size has been incresed then Sales have been incresed.

```python
#Correlation Analysis
sns.heatmap(df[["OperatingCost","Staff_Cnt","TotalSales","Total_Customers","AcqCostPercust","BasketSize","ProfitPercust","OwnStore","OnlinePresence","Tenure","StoreSegment"]].corr(),annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
